# Loan Price prediction

```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("loanpred_train.csv")
df.head()
```

```python
df['Loan_Status'] = df['Loan_Status'].map({'Y':1 ,'N':0})
```

```python
df.head(10)
```

```python
X = df.drop(['Loan_ID', 'Loan_Status'], axis = 1)
y = df.Loan_Status
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.30,random_state =123)
```

```python
lp = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = lp.fit(X_train, X_test, y_train, y_test)
```

```python
print(models)
```

# Titanic Dataset

```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
X=df.drop('survived',axis=1)
y=df.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.25,random_state =123)
```

```python
lp_td = LazyClassifier(verbose=0, ignore_warnings=False, custom_metric=None)
models,predictions = lp_td.fit(X_train, X_test, y_train, y_test)
```
